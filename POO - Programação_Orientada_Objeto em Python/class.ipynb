```python
class Dog:
    def __init__(self, raca):
        self.raca = raca
        self.idade = 10
        print("{} criado".format(self.raca))
    
    def envelhecer(self):
        self.idade += 1
```

```python
dog = Dog("Labrador")
dog2 = Dog("Poodle")
```

```python
dog.envelhecer()
```

```python
print(dog.idade)
print(dog.raca)
```

```python
print(dog2.idade)
print(dog2.raca)
```

## Circulo


```python
class circulo:
    def __init__(self, raio=1):
        self.raio = raio

    def calcular_area(self):
        return self.raio * self.raio * 3.14
    
    def retorna_raio(self):
        return self.raio
```

```python
c1 = circulo()
c2 = circulo(4)
(c1.raio)
```

```python
(c2.raio)
```

```python
c1.calcular_area()
```

```python
c2.calcular_area()
```

```python
c1.retorna_raio()
```

```python
c2.retorna_raio()
```
