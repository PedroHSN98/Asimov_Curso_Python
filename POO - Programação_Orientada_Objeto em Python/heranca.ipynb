## Herança


```python
class Animal:
    def __init__(self):
        print('Animal Criado')

    def quem_sou_eu(self):
        print('Eu sou um Animal')
    
    def comer(self):
        print('Eu estou comendo')

class Cachorro(Animal):
    def __init__(self):
        Animal.__init__(self)
        print('Eu sou um cachorro')
    def quem_sou_eu(self):
        print('Eu sou um Cachorro')
```

```python
animal = Animal()
```

```python
animal.quem_sou_eu()
```

```python
dog = Cachorro()
```

```python
dog.quem_sou_eu()
```
