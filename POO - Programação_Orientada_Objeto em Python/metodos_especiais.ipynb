## Metodos Especiais

```python
class Book(object):
    def __init__(self, title, author, pages):
        print("A book is being created")
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return "Title:%s, Author:%s, Pages:%s" % (self.title, self.author, self.pages)

    def __len__(self):
        return self.pages

    def __del__(self):
        print("A book object has been deleted")
```

```python
book = Book("Python Rocks!", "Jose Portilla", 159)
#Metodos Especiaos
print(book)
print(len(book))
del book
```
