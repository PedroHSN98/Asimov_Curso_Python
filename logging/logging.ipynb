### Logging


```python
import logging
```

```python
LOG_FORMAT = '%(levelname)s %(asctime)s - %(message)s'
```

```python
logging.basicConfig(filename='app.log', level=logging.DEBUG, filemode='w', format=LOG_FORMAT)
```

```python
log = logging.getLogger()
```

```python
log.info('Olá, mundo!')
log.critical('Critico')
log.error('Error')
log.warning('Warning')
```
