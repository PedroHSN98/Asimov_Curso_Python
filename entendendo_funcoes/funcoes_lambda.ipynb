### Funções Lambda

```python
def meu_filtro(x):
    return x > 2


filtro = filter(lambda x: x > 2, [1, 2, 3, 4])
list(filtro)
```

```python
def func(a, b, c):
    return (a + b) > c

func_2 = lambda a, b, c: (a + b) > c
```

```python
func_2(1, 2, 3)
```

```python
list(filter(lambda x: len(x) <= 3, ['samasndasdad', 'sad', '']))
#mapa = list(map(lambda x: str(x + 2), [1, 2, 3]))
#list(mapa)
```

```python
lista = ['abc', 1, 2, 5.3, '.', '22']
lista_ordenada = sorted(lista, key=lambda x: len(x) if isinstance(x, str) else x)
```
