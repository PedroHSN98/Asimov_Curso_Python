### Funções Built-in

```python
#isintance
valor_1 = 'Python'
valor = 2
if isinstance(valor_1, (int, float)):
    print('valor é do tipo numerico')
else:
    print('valor não é do tipo numerico')
```

```python
#any e all
booleanos = [True, False, True]
all(booleanos)
```

```python
any(booleanos)
```

```python
valores = [1, 2, 3, 3.2, 4, 5]

if all(isinstance(valor, int) for valor in valores):
    print('Todos os valores são int')
else: 
    print('Nem todos os valores são int')
```

```python
#Map
def somar_dois(n):
    return n + 2

numeros = [3, 6, 10]
mapa = map(somar_dois, numeros)
```

```python
list(mapa)
```

```python
numeros_mais_dois = [somar_dois(n) for n in numeros]
numeros_mais_dois
```

```python
#filter
def meu_filtro(n):
    return n > 5

numeros = [3, 6, 10]
filtro = filter(meu_filtro, numeros)
```

```python
list(filtro)
```

```python
maiores_que_cinco = [n for n in numeros if meu_filtro(n)]
maiores_que_cinco
```
