# Itertools

```python
#itertools.chain
import itertools

seq1 = (1, 2, 3)
seq2 = ['a', 'b', 'c']

for elemento in itertools.chain(seq1, seq2, seq2):
    print(elemento)
```

```python
#Itertools.zip_longest
nomes = ['Juliano', 'José', 'Lucas', 'Luiza']
idade = [30, 25, 28]
cpfs = ['xxx', 'yyy', 'zzz']
emails = ['juliano@empresa.com', 'laura@empresa.com']

for elemento in itertools.zip_longest(nomes, idade, cpfs, emails):
    print(elemento)
```

```python
#itertools.product
comidas = ['Pizza', 'Hambúrguer', 'Sushi']
bebidas = ['Coca-cola', 'Suco']

for elemento in itertools.product(comidas, bebidas):
    print(elemento)
```

```python
#itertools.combinations
nomes = ['Juliano', 'José', 'Lucas', 'Luiza']

for comb in itertools.combinations(nomes, 2):
    print(comb)
```
