```python
dic = {
    'chave1' : 'valor1',
    'chave2' : 'valor2',
    'chave3' : 'valor3',
}

for chave, valor in dic.items():
    print(f'Chave: {chave} -> Valor: {valor}')
```

```python
nomes = ['Juliano', 'José', 'Lucas', 'Luiza']
idade = [30, 25, 28, 22]
```

```python
for elemento in enumerate(zip(nomes, idade)):
    print(elemento)
```

```python
for i, (nome, idade) in enumerate(zip(nomes, idade)):
    print(i, nome, idade)
```
