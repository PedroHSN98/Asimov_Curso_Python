## Operadores de Divisão
# O Operador piso(divisão inteira)                    
O operador *//* retorna a divisão sem a parte fracionaria.

# Operador Módulo(resto da divisão)
O Operador % retorna o "resto" de uma divisão, em unidades inteiras


```python
10 / 3
```

```python
10 // 3
```

```python
n = 10 % 3
print(n)
print(11 % 3)
```

```python
segundos_totais = 32451
segundos_totais / 60
```

```python
minutos = segundos_totais // 60
print(minutos)
```

```python
segundos = segundos_totais % 60
segundos
```

```python
print(f'{minutos} minutos e {segundos} segundos')
```

```python
n = 2
div = 2
par = n % 2 == 0
print(f'O Número {n} é par? {par}')
```

```python
n = 18
div = 9
divisor = (n % div) == 0
print(f'O Número {div} é divisor de {n}? {divisor}')
```
