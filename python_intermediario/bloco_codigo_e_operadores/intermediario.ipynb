## O bloco With


```python
arquivo = open('notas.txt', 'w')
arquivo.write('Esta é uma anotação especial')

arquivo.close()
```

```python
with open('notas.txt', 'w') as arquivo:
    arquivo.write('Esta é uma anotação especial')

print('Arquivo foi fechado!')
```
