# Expressão Condicional

```python
x = 4
y = 5

if x > y:
    maior_valor = x
else:
    maior_valor = y

maior_valor 
```

Operador Ternario ( outras linguagens)
VAR = CONDIÇÃO ? VALOR DE VERDADEIRO ; VALOR SE FOR FALSO

Expressão Condicional (Python)
VAR = VALOR_VERDADEIRO SE CONDIÇÃO SENÃO VALOR_FALSO

```python
maior_valor = x if x > y else y
maior_valor
```

```python
def pegar_cor(valor):
    return "vermelho" if valor < 0 else "azul" 


for valor in [-1, 0, 1]:
    print(f'A cor do valor {valor} é {pegar_cor(valor)}')
```

```python
numeros = [1, 2, 3, 4, 5]
pares_quadrados = [
    n ** 2
    if n % 2 == 0 
    else n 
    for n in numeros
] 

pares_quadrados
```
