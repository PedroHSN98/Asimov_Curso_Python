### Usando type hints - tipagem de dados em Python

```python
x: int = 10
y: float = 5.5
nome : str = 'Juliano'

nomes: list[str] = ['Luiza,' 'Henrique', 'Ricavalores']

valores: tuple[int] = (1, 2, 3, 4)

produtos: dict[str, float] = {
    'carne': 4.4,
    'leite': 2.1,
}

```
