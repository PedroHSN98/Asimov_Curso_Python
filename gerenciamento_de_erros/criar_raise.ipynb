### Criando nossas proprias exceções com raise


```python
import math

def calcular_raiz_quadrada(numero):
    if numero < 0:
        raise ValueError(f'Impossivel de calcular raiz quadrada de numero negativo!')
    return math.sqrt(numero)

for n in [4, 2, 1, 0, -1]:
    raiz = calcular_raiz_quadrada(n)
    print(f'Raiz quadrada de {n} é {raiz}')
```
